La demostracion es: $P(k+1)=\frac{\lambda}{k+1}P(k)$, hay que definir las siguientes propiedades de los factoriales:$$1. k!=k(1+k)!$$$$2. \frac{k!}{(k+1)!}=\frac{1}{k+1}$$

El resultado es:$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$

$$\frac{P(k+1)}{P(k)} = \frac{e^{-\lambda}\lambda^{k+1} k!}{(k+1)!e^{-\lambda}\lambda^k} = \frac{e^{-\lambda}\lambda^{k}\lambda k!}{(k+1)!e^{-\lambda}\lambda^k}$$$$\frac{P(k+1)}{P(k)} = \frac{\lambda k!}{(k+1)!}$$$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1}$$$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1} = \frac{1}{(k+1)} * \frac{\lambda}{1}=\frac{\lambda}{(k+1)}$$$$P(k+1) = \frac{\lambda}{(k+1)}P(k)$$

```python
def P_acum_Poisson_o(k,l):
    global Pr
    Pr = np.zeros(k)
    Pr[0] = np.exp(-l)
    def pr(i):
        global Pr
        Pr[i+1]=l/(i+1)*Pr[i]
        
    # Lleno el vector Pr usando compresión de listas
    [pr(i) for i in range(k-1)]
    return np.cumsum(Pr)
n = 12; l = 4; N = 10**4

p_acum = P_acum_Poisson_o(n,l)
u = np.random.rand(N)

# Gráfica de resultados
plt.figure(figsize=[13,4])
plt.subplot(121)
d_pois = list(map(lambda j:sum(1 for i in p_acum if i<u[j]),range(N)))
plt.hist(d_pois,100,density=True,label='Distribucion de Possion')
plt.legend()

```

$$ c =\int_{0}^{1} t(x) dx =\int_{0}^{1} a \sin(\pi x) dx  $$$$ a\int_{0}^{1} \sin(\pi x) dx = a[\frac{-\cos(\pi x)}{\pi}]_{0}^{1} $$$$ a[\frac{-\cos(\pi x)}{\pi}]_{0}^{1} = \frac{a}{\pi} - (- \frac{a}{\pi}) = \frac{2a}{\pi} $$$$ c = \frac{2a}{\pi} $$

Obtenemos:$$ g(x) = \frac{\pi \sin(\pi x)}{2}$$

$$\frac{p_{i+1}}{p_i} = \frac{n!p^ip(1-p)^ni!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!n!p^i(1-p)^n}$$$$\frac{p_{i+1}}{p_i} = \frac{pp^i(1-p)^ni!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!p^i(1-p)^n} = \frac{pi!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!}=\frac{pi!(n-i)!}{(1-p)(i+1)!(n-i-1)!}$$

La demostracion es: $P(k+1)=\frac{\lambda}{k+1}P(k)$, hay que definir las siguientes propiedades de los factoriales:$$1. k!=k(1+k)!$$$$2. \frac{k!}{(k+1)!}=\frac{1}{k+1}$$

El resultado es:$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$

$$\frac{P(k+1)}{P(k)} = \frac{e^{-\lambda}\lambda^{k+1} k!}{(k+1)!e^{-\lambda}\lambda^k} = \frac{e^{-\lambda}\lambda^{k}\lambda k!}{(k+1)!e^{-\lambda}\lambda^k}$$$$\frac{P(k+1)}{P(k)} = \frac{\lambda k!}{(k+1)!}$$$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1}$$$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1} = \frac{1}{(k+1)} * \frac{\lambda}{1}=\frac{\lambda}{(k+1)}$$$$P(k+1) = \frac{\lambda}{(k+1)}P(k)$$



```python
###Se crea una función que genera la función de distribución de Poisson acumulada

def poisson(lam:"lamda",n:"n"): 
  
    vector = np.zeros(n)
    vector[0] = (np.exp(-lam)) 
    def fill_vector(i):
        nonlocal vector
        vector[i+1] = (lam*vector[i])/(i+1) 
    [fill_vector(i) for i in range(n-1)]
    return vector.cumsum()



def distr_discreta(p_acum: 'P.Acumulada',
                       N: 'numero a generar'):
    
    U =np.random.rand(N)
    v = [sum([1 for p in p_acum if p < ui ])  for ui in U]
    return v


def histogram(distribucion:'dist del hist',
                            label:'label'):
    plt.figure(figsize=[8,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion))) 
    plt.bar(list(set(distribucion)),y,label=label)
    plt.legend()
    plt.show()
    
n,lam= 1000,4 
p_acum = poisson(lam,n) 
N=1000 
distribucion = distr_discreta(p_acum,N) 
histogram(distribucion,"Distribución de Poisson") 
```
