$g(x)$ está definida como $g(x) = \frac{t(x)}{c}$. Tenemos que $t(x)=a \sin(\pi x)$ y $c$ como $c =\int_{0}^{1} t(x) dx$. Entonces se pude operar de la siguiente manera:

$$ c =\int_{0}^{1} t(x) dx =\int_{0}^{1} a \sin(\pi x) dx  $$$$ a\int_{0}^{1} \sin(\pi x) dx = a[\frac{-\cos(\pi x)}{\pi}]_{0}^{1} $$$$ a[\frac{-\cos(\pi x)}{\pi}]_{0}^{1} = \frac{a}{\pi} - (- \frac{a}{\pi}) = \frac{2a}{\pi} $$$$ c = \frac{2a}{\pi} $$


$$g(x) = \frac{a \sin(\pi x)}{\frac{2a}{\pi}} = \frac{\frac{a \sin(\pi x)}{1}}{\frac{2a}{\pi}} = \frac{a\pi \sin(\pi x)}{2a} = \frac{\pi \sin(\pi x)}{2}$$
Obtenemos:$$ g(x) = \frac{\pi \sin(\pi x)}{2}$$

```python
# Ejercicio 5:

# Esta funcion regresa los terminis aceptados y rechazados
def yes_no(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo',
                 t:'función mayor f'):

    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'om') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'oc') 
     for i in range(len(R1))]
                
    plt.show()
    
    x = [R1[i] for i in range(len(R1)) if condition[i]]
    return x


# Grafica
plt.figure(figsize=[8,6])
plt.title('Gráfica de $g(x)$ y su distribución',fontsize=18)
f = lambda x: 30*(x**2 - 2*x**3 + x**4) 
a = f(optimize.fmin(lambda x:-f(x), 0, disp=False)) 
x =np.arange(0,1,0.01)
t = lambda x: a * np.sin(np.pi*x)
plt.plot(x,f(x), label='$f(x)=30(x^2-2x^3+x^4)$') 
plt.plot(x,t(x),label='$t(x)=a *\sin(\pi x)$')
plt.legend(loc="best")


# simulacion del metodo
N = 500 
R2 = np.random.rand(N) 
R1 = np.random.uniform(0,1, size=N)
x_s = yes_no(R2,R1,f,t)
```

Tal cómo se puede ver graficamente la función $t(x)$ muy cercana al rango de la función $f(x)$ permite acotar de manera más precisa los números aleatorios que se meten al algoritmo, al rededor del 70% de aceptacion tienen

L demostracion es: $p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$, hay que definir las siguientes propiedades de los factoriales:$$1. n!=n(1+n)!$$y$$2. \frac{n!}{(n+1)!}=\frac{1}{n+1}$$

sabiendo que:$$ \frac{p_{i+1}}{p_i}=\frac{\frac{n! p ^ {i+1} (1-p)^{n-i+1}}{(i+1)!(n-i+1)!}}{\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i}}$$

$$\frac{p_{i+1}}{p_i} = \frac{n!p^ip(1-p)^ni!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!n!p^i(1-p)^n}$$$$\frac{p_{i+1}}{p_i} = \frac{pp^i(1-p)^ni!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!p^i(1-p)^n} = \frac{pi!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!}=\frac{pi!(n-i)!}{(1-p)(i+1)!(n-i-1)!}$$
$$\frac{p_{i+1}}{p_i} = \frac{pi!(n-i)!}{(1-p)i!(i+1)(n-i-1)!} = \frac{p(n-i)!}{(1-p)(i+1)(n-i-1)!}$$$$\frac{p_{i+1}}{p_i} = \frac{p(n-i)}{(1-p)(i+1)} = \frac{n-i}{i+1}\frac{p}{1-p}$$$$p_{i+1} = \frac{n-i}{i+1}\frac{p}{1-p}p_i$$


Las ventajas de que sea recursiva nos ayudan a programar a través de vectores el comportamiento de dicha sucesión, ya que de no contar con está recursividad se vuelve compleja la programación de los factoriales.


# 9

La demostracion es: $P(k+1)=\frac{\lambda}{k+1}P(k)$, hay que definir las siguientes propiedades de los factoriales:$$1. k!=k(1+k)!$$$$2. \frac{k!}{(k+1)!}=\frac{1}{k+1}$$

El resultado es:$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$

$$\frac{P(k+1)}{P(k)} = \frac{e^{-\lambda}\lambda^{k+1} k!}{(k+1)!e^{-\lambda}\lambda^k} = \frac{e^{-\lambda}\lambda^{k}\lambda k!}{(k+1)!e^{-\lambda}\lambda^k}$$$$\frac{P(k+1)}{P(k)} = \frac{\lambda k!}{(k+1)!}$$$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1}$$$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1} = \frac{1}{(k+1)} * \frac{\lambda}{1}=\frac{\lambda}{(k+1)}$$$$P(k+1) = \frac{\lambda}{(k+1)}P(k)$$



```python
###Se crea una función que genera la función de distribución de Poisson acumulada

def poisson(lam:"lamda",n:"n"): 
  
    vector = np.zeros(n)
    vector[0] = (np.exp(-lam)) 
    def fill_vector(i):
        nonlocal vector
        vector[i+1] = (lam*vector[i])/(i+1) 
    [fill_vector(i) for i in range(n-1)]
    return vector.cumsum()



def distr_discreta(p_acum: 'P.Acumulada',
                       N: 'numero a generar'):
    
    U =np.random.rand(N)
    v = [sum([1 for p in p_acum if p < ui ])  for ui in U]
    return v


def histogram(distribucion:'dist del hist',
                            label:'label'):
    plt.figure(figsize=[8,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion))) 
    plt.bar(list(set(distribucion)),y,label=label)
    plt.legend()
    plt.show()
    
n,lam= 1000,4 
p_acum = poisson(lam,n) 
N=1000 
distribucion = distr_discreta(p_acum,N) 
histogram(distribucion,"Distribución de Poisson") 
```
